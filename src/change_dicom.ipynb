```python
from PIL import Image

def flip_l_to_r(path):
    os.chdir(path)
    files = [f for f in os.listdir() if 'LEFT' in f]
    
    # Loop through files and mirror flip them
    for f in files:
        image = Image.open(f)
        array = np.fliplr(np.asarray(image))
        plt.imshow(array)
        plt.savefig(f, dpi=350)
    
    return 0 
```
